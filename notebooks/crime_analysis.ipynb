#### Incidents Observations

- Grain of dataset: Each row represents one crime incident reported in California (2024). No aggregation is done at row level.

- Key identifier: incident_id (used to uniquely identify each incident and link with other tables)

- Time-related columns:
  - submission_date → when the record was submitted (string, needs datetime conversion)
  - incident_date → when the crime happened (string, needs datetime conversion)
  - incident_hour → hour of incident (partially missing ~27,036 values)
  - nibrs_month_id → encoded month identifier
  - cleared_except_date → date of case clearance (very sparse)

- Missing values:
  - incident_hour → missing for some records (unknown time)
  - cleared_except_date → mostly missing (cases not cleared or not updated)
  - data_home → completely empty (no usable information)

- Column types issue:
  - Dates are stored as strings (not datetime yet)
  - Most IDs are integers (used for linking, not calculation)

- System/unused fields:
  - data_home → empty column (no data present)
  - orig_format → data source format indicator (not used for analysis)
  - did → internal system identifier (not needed for analysis)

#### Victims Observations

- Grain of dataset: Each row represents one victim record linked to a specific incident in 2024. One incident can have multiple victim records.

- Key relationship: incident_id (used to connect victim data with the main incidents table)

- Multiple victims per incident: Yes. This is supported by repeated incident_id values and victim_seq_num indicating multiple victims within the same incident.

- Important attributes:
  - victim_type_id (type/category of victim)
  - age_num (victim age, stored as string)
  - sex_code (gender of victim)
  - race_id, ethnicity_id (demographic information)
  - resident_status_code (residence status, partially missing)

- Missing patterns:
  - Very high missing values in assignment_type_id, activity_type_id, and outside_agency_id (mostly unused in this dataset)
  - resident_status_code is missing for many records (~50%)
  - age_range_low_num partially missing
  - age_code_range_high mostly missing

- Use:
  - Study victim demographics (age, sex, race, ethnicity)
  - Analyze victim distribution across incidents
  - Compare victim characteristics across crime types (after merging with incidents)

#### Offenders Observations

- Grain: Each row represents one offender involved in a crime incident. One incident can have multiple offenders.

- Relationship to incidents: Linked using incident_id. This is a many-to-one relationship (many offenders per incident).

- Key fields:
  - offender_id → unique identifier for each offender record
  - incident_id → links offender to incident dataset
  - offender_seq_num → order of offenders within the same incident
  - age_num → offender age (string format, more reliable than range fields)
  - sex_code, race_id, ethnicity_id → demographic attributes

- Missing data patterns:
  - age_range_low_num is missing for ~614,290 records
  - age_range_high_num is missing for ~1,265,469 records
  - age range fields are highly incomplete and not reliable for most records
  - core identifiers and demographic fields are fully populated

- Use:
  - Study offender demographics (age, sex, race, ethnicity)
  - Analyze number of offenders per incident
  - Compare offender profiles across different incident types (after merging with incidents dataset)

#### Agencies Observations

- Grain: Each row represents one law enforcement agency for the year 2024.

- Key field: agency_id (used to link with the Incidents dataset).

- Useful attributes:
  - ucr_agency_name / pub_agency_name → agency name
  - agency_type_name → type of law enforcement agency
  - agency_status → agency’s operational status
  - county_name, msa_name, state_name → agency location
  - population → population served by the agency
  - reporting_type, participated, nibrs_participated → reporting and participation information

- Missing data patterns:
  - covered_by_legacy_ori is almost completely empty.
  - dormant_year is completely empty.
  - pub_agency_unit and some staffing-related columns contain many missing values.
  - Most important identification and location columns have no missing values.

- Use:
  - Join with the Incidents dataset using agency_id.
  - Identify which agency handled each incident.
  - Compare crime across different agencies, counties, or metropolitan areas.
  - Add agency information (name, location, population, etc.) to incident-level analysis.

## Data Ingestion Summary
- Incidents dataset is the central table for analysis.
- Victims, offenders, and arrestees link to incidents using incident_id (one-to-many relationship).
- Agencies link to incidents using agency_id.
- Offenses provide details of crimes within each incident (multiple per incident).
- Lookup tables are used to decode coded fields, not for direct analysis.
- Many columns contain missing values and will need cleaning before analysis.
- Overall dataset is relational and requires joins across multiple tables for analysis.
